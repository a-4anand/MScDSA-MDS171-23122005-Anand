Create a Python Class for managing student details and marks.
Define the class and implement the methods of the student class in a menu-driven program for different types of users

```python
class StudentManager:
    def __init__(self):
        self.students={}

    def addDetails(self,name,roll_number,marks):
        if roll_number not in self.students:
             self.students[roll_number]={"name":name,"marks":marks}
             print(f"The details for {name} has been added succsesfully ")
        else:
            print("the details for this student aready exist")

    def modifyDetails(self,name,roll_number,marks):
        if roll_number in self.students:
             self.students[roll_number]["name"]=name
             self.students[roll_number]["marks"]=marks

        print(f"The details for {name} has been updated succesfully")
    
    def veiwData(self,roll_number):
        if roll_number in self.students:
            student=self.students[roll_number]
            print(f"Roll Number : {roll_number}")
            print(f"Name : {student['name']}")
            print(f"Marks : {student['marks']}")
        else:
            print("No details avialable for this roll number")
    
    def deleteData(self,roll_number):
        if roll_number in self.students:
            del self.students[roll_number]
            print(f"Details for roll number {roll_number} have been deleted.")
        else:
            print("No details available for this roll number.")

student_manager = StudentManager()

while True:
    
    print("="*40)
    print("Welcome To Alwin Convent School")
    print("="*40)
    print("1.Veiw Details")
    print("2.Add Details")
    print("3.Modify Details")
    print("4.Delete Data")
    print("5.Exit")
    print("="*40)

    choice=input("Please Enter Your choice: ")

    if choice == "1":
        roll_number = input("Enter Roll Number: ")
        student_manager.veiwData(roll_number)
    elif choice == "2":
        name=input("Please Enter the name: ")
        roll_number=input("Please Enter the roll number: ")
        marks=input("Please Enter the marks: ")
        student_manager.addDetails(name,roll_number,marks)
    elif choice == "3":
        name=input("Please Enter the New name: ")
        roll_number=input("Please Enter the New roll number: ")
        marks=input("Please Enter the New marks: ")
        student_manager.modifyDetails(name,roll_number,marks)
    elif choice =="4":
        roll_number = input("Enter Roll Number: ")
        student_manager.deleteData(roll_number)
    elif choice =="5":
        print("exiting The program")
        break
    else: 
        print("Invalid Input")





```
